```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv('/content/train.csv')
submission = pd.read_csv('/content/sample_submission.csv')

# 'timestamp' 열이 없으면 추가
if 'timestamp' not in train.columns:
    timestamp_col = train.columns[0]  # 실제 데이터프레임에서 사용할 열의 이름으로 수정해주세요
    train['timestamp'] = pd.to_datetime(train[timestamp_col]).dt.date
    submission['timestamp'] = pd.to_datetime(submission[timestamp_col]).dt.date

# 'timestamp' 열 제거
train = train.drop(columns=['timestamp'])
submission = submission.drop(columns=['timestamp'])

# "train" 데이터 생성 (예제)
train['input_feature'] = range(len(train))  # 예제로 생성한 입력 피처, 실제 데이터에 따라 다르게 생성

# "submission" 데이터 생성 (예제)
submission['input_feature'] = range(len(submission))  # 예제로 생성한 입력 피처, 실제 데이터에 따라 다르게 생성

# 훈련 데이터와 테스트 데이터로 분리
train_input, _, train_target, _ = train_test_split(train, submission, test_size=0.3, random_state=42)

# 다항 특성 추가
poly = PolynomialFeatures()
train_poly = poly.fit_transform(train_input)
submission_poly = poly.transform(submission)

# 표준화
ss = StandardScaler()
train_scaled = ss.fit_transform(train_poly)
submission_scaled = ss.transform(submission_poly)

# 선형 회귀 모델 훈련
lr = LinearRegression()
lr.fit(train_scaled, train_target)

# 예측
submission_predictions = lr.predict(submission_scaled)

# 결과 확인
print(submission_predictions)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv('/content/train.csv')
submission = pd.read_csv('/content/sample_submission.csv')

# "train" 데이터 생성 (예제)
train['input_feature'] = range(len(train))  # 예제로 생성한 입력 피처, 실제 데이터에 따라 다르게 생성

# "submission" 데이터 생성 (예제)
submission_expanded = pd.concat([submission] * len(train), ignore_index=True).head(len(train))
submission_expanded['input_feature'] = range(len(submission_expanded))

# 'timestamp' 및 'date' 열 제거
train = train.select_dtypes(include=[np.number])  # 숫자형 열만 선택
submission_expanded = submission_expanded.select_dtypes(include=[np.number])

# NaN 값을 해당 열의 평균값으로 대체 (SimpleImputer 사용)
imputer = SimpleImputer(strategy='mean')
train_imputed = pd.DataFrame(imputer.fit_transform(train), columns=train.columns)
submission_imputed = pd.DataFrame(imputer.transform(submission_expanded), columns=submission_expanded.columns)

# 훈련 데이터와 테스트 데이터로 분리
train_input, test_input, train_target, test_target = train_test_split(train_imputed, submission_imputed, test_size=0.3, random_state=42)

# 다항 특성 추가
poly = PolynomialFeatures()
train_poly = poly.fit_transform(train_input)
test_poly = poly.transform(test_input)

# 표준화
ss = StandardScaler()
train_scaled = ss.fit_transform(train_poly)
test_scaled = ss.transform(test_poly)

# PolynomialFeatures에서 생성된 특성 이름을 가져옴
poly_feature_names = poly.get_feature_names(train.columns)

# 표준화된 데이터를 데이터프레임으로 변환 (이때 특성 이름을 지정)
train_scaled_df = pd.DataFrame(train_scaled, columns=poly_feature_names)
test_scaled_df = pd.DataFrame(test_scaled, columns=poly_feature_names)

# 선형 회귀 모델 훈련
lr = LinearRegression()
lr.fit(train_scaled_df, train_target)

# 모델 평가
score_train = lr.score(train_scaled_df, train_target)
score_test = lr.score(test_scaled_df, test_target)

print("R^2 Score (Train):", score_train)
print("R^2 Score (Test):", score_test)
```
