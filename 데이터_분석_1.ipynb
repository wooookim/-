```python
with open('남산도서관 장서 대출목록 (2021월 04월).csv') as f:
  print(f.readline())
```

파이썬 open 함수는 기본적으로 텍스트 파일이 utf-8 형식 저장을 가정

- open 함수를 열 때 mode 매개변수를 바이너리 읽기모드(rb)로 지정하면 인코딩 형식에 상관 없이 파일을 열 수 있음

```python
import chardet
with open('남산도서관 장서 대출목록 (2021월 04월).csv', mode = 'rb') as f:
  d = f.readline()
print(chardet.detect(d))
```
