```python
import gdown
gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet = False)
```

```python
import pandas as pd
ns_book7 = pd.read_csv('ns_book7.csv', low_memory = False)
ns_book7.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
plt.show()
```

```python
'''
figsize -> 실제 사이즈를 인치 단위로 변경
'''
plt.figure(figsize = (9, 6))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
plt.show()
```

```python
plt.figure(figsize = (900/72, 600/72))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
plt.show()
```

```python
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
plt.figure(figsize = (900/72, 600/72))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
plt.show()
```

```python
%config Inlinebackend.print_figure_kwargs = {'bbox_inches': 'tight'} # 그래프 주변 공백 최소화 / 코랩 기본 설정
```

```python
plt.figure(dpi = 144)   # 픽셀 크기 조정
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
plt.show()
```

```python
'''
rcParams 객체
- 맷플롯립 그래프 기본값을 관리하는 객체
'''
plt.rcParams['figure.dpi'] = 100
```

```python
plt.rcParams['scatter.marker']
```

```python
plt.rcParams['scatter.marker'] = '*'
```

```python
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
plt.show()
```

```python
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1, marker = '+')
plt.show()
```

```python
'''
서브플롯
- 맷플롯립의 Axes 클래스의 객체
- 2개 이상의 축을 포함
'''
fig, axs = plt.subplots(2)

# 첫번째 그래프
axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)

# 두번째 그래프
axs[1].hist(ns_book7['대출건수'], bins = 100)
axs[1].set_yscale('log')

fig.show()   # plt와 혼동 주의
```

```python
fig, axs = plt.subplots(2, figsize = (6, 8))

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
axs[0].set_title('scatter plot')

axs[1].hist(ns_book7['대출건수'], bins = 100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')

fig.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 4))  # 행, 열, 크기

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
axs[0].set_title('scatter plot')
axs[0].set_xlabel('number of books')
axs[0].set_ylabel('borrow count')

axs[1].hist(ns_book7['대출건수'], bins = 100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')
axs[1].set_xlabel('borrow count')
axs[1].set_ylabel('frequency')

fig.show()
```
